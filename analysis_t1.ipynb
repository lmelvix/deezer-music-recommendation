## Temporal Analysis

```python
# Hour of the day
data['ts_listen_hour'] = data['ts_listen']/3600
data['ts_listen_hour'] = data['ts_listen_hour'].apply(np.floor)
data['ts_listen_hour'] = data['ts_listen_hour']%24


s = data.groupby(['ts_listen_hour'])['is_listened'].mean()
plt.figure(figsize=(20,10))
plt.bar(s.index, s.values, color = 'maroon', alpha=0.75)
plt.xlabel('Hour of the Day', fontsize=30)
plt.ylabel('Number of Songs', fontsize=30)
```

```python
# Incorrect data removal. Computation of listen_time-release_time
df = data[(data['release_date'] < 20180000) & (data['ts_listen'] > 10000)]
s1 = pd.to_datetime(df['ts_listen'], unit='s').dt.date
s2 = pd.to_datetime(df['release_date'], format='%Y%m%d').dt.date
s = s1 - s2
#data['listen_release_diff'] = (datetime.fromtimestamp(data['ts_listen'])  - datetime.strptime(data['release_date'], '%Y%m%d'))
# data['listen_release_diff'] = data['listen_release_diff'].apply(np.floor).apply(np.int)
```

```python
df['listen_release_diff'] = s
```

```python
df
```

```python
df['listen_release_diff_month'] = df['listen_release_diff'].map(lambda x: round(x.days/30))
```

```python
# Listen count with days
s = df.groupby(['listen_release_diff'])['is_listened'].count()
plt.figure(figsize=(20,10))
plt.plot(s.index,s.values, color = 'maroon', alpha=0.75)
plt.xlabel('Days from Release', fontsize=30)
plt.ylabel('Songs Listened', fontsize=30)
```

```python
# Mean by month.
s = df.groupby(['listen_release_diff_month'])['is_listened'].mean()
plt.figure(figsize=(20,10))
plt.plot(s.index,s.values, color = 'maroon', alpha=0.75)
plt.xlabel('Months from Release', fontsize=30)
plt.ylabel('Number of Songs', fontsize=30)
```
